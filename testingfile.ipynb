```python
import numpy as np

def npz_viewer(npz_file="en_speaker_6.npz"):
    data = np.load(npz_file)
    for key in data.files:
        print(f"Variable: {key}")
        print(data[key])

# Provide a default file if one is not specified in the notebook cell
npz_viewer()
```
